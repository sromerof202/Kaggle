```python
# Importing necessary libraries
import pandas as pd
import statsmodels.api as sm

# Kaggle Dataset File Path
data_file_path = '/kaggle/input/ai-chatbots-perceptions-studentsstafffaculty/Perceptions_towards_AI_Survey.csv'

# Load your data
data = pd.read_csv(data_file_path)

# Selecting UTAUT columns (assuming same column names as in the previous analysis)
utaut_columns = [col for col in data.columns if "11. Questions/Statements" in col]
utaut_data = data[utaut_columns]

# Mapping Likert scale responses to numerical values
likert_mapping = {
    'Strongly Disagree': 1,
    'Disagree': 2,
    'Neutral': 3,
    'Agree': 4,
    'Strongly Agree': 5
}
utaut_data_numerical = utaut_data.replace(likert_mapping)

# Separating independent and dependent variables
X = utaut_data_numerical.iloc[:, :-2]  # Independent variables
y = utaut_data_numerical.iloc[:, -2:].mean(axis=1)  # Dependent variable (average of BI related columns)

# Adding a constant to the model (for intercept)
X = sm.add_constant(X)

# Fitting the OLS model
model = sm.OLS(y, X).fit()

# Getting the summary of the regression model
model_summary = model.summary()

# Displaying the summary
print(model_summary)

```
